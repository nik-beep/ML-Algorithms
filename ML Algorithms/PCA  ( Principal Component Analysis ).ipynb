# Dimensionality Reduction  

```python
# Import Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
# Load Data Set 
os.chdir('C:\\Users\\nikhil reddy\\Downloads\\Day20\\Day20\\')
df = pd.read_csv('trans_us.csv', index_col = 0, thousands  = ',')
df
```

```python
# Update Row and Column Headings 
df.index.names = ['stations']
df.columns.names = ['Years']
df
```

```python
# Check for Null Values 
df.isna().sum()
```

```python
# Create Model 
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)
```

```python
# Create PCA  with two variables 
PCA1 = pca.transform(df)
PCA1
```

```python
# Convert output to Data Frame 
PCA2= pd.DataFrame(PCA1)
PCA2.index = df.index
PCA2.columns = ['PC1','PC2']
PCA2.head(50)
```

```python
# Display the variance Percentage  
pca.explained_variance_ratio_
```

```python
os.chdir ('C:\\Users\\nikhil reddy\\Downloads\\Day20\\Day20\\')
df= pd.read_csv('Bank Data Dimension Reduction.csv')
df
```

```python
# Data Checks 
df.isna().sum()
```

```python
df.notna().sum()
```

```python
# Data Types 
df.dtypes
```

```python
df.groupby('Gender').size()
```

```python
df['Gender'] = df['Gender'].fillna(0)
df.groupby('Gender').size()
```

```python
df.groupby('Previous').size()
df['Previous'] = df['Previous'].fillna(0)
```

```python
df['Interest_Rate'].fillna(df['Interest_Rate'].mean(),inplace = True)
```

```python
df['Duration'].fillna(df['Duration'].median(),inplace = True)
```

```python
df['Gender'] = df['Gender']. astype('category')
df.dtypes
```

```python
df['Previous'] = df['Previous']. astype('category')
df.dtypes
```

```python
df.dtypes
df['Churn'] = df['Churn']. astype('category')
df['Churn'] = df['Churn'].cat.codes
```

```python
x = df.iloc[:,1:6].values
x
```

```python
y = df.iloc[:,6].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x = sc_x.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Display X Train as Data Frame 
pd.DataFrame(x_train)
```

Apply PCA model in x_train and x_test 
 n_components = None ( Maximum Slicing , Slicing based on number of variable in this case maximum slicing is 5 – Number of Independent variables)


```python
from sklearn.decomposition import PCA
pca = PCA(n_components = None)
x_train = pca.fit_transform( x_train)
x_test = pca.transform(x_test)
```

```python
# Display X Train as Data Frame 
pd.DataFrame(x_train)
```
